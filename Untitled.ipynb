```python
!pip install feature-engine
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import sklearn

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from feature_engine.datetime import DatetimeFeatures
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import learning_curve
```

```python
pd.set_option("display.max_columns",None)
```

```python
sklearn.set_config(transform_output='pandas')
```

```python
train_df = pd.read_csv('/content/train.csv')
test_df = pd.read_csv('/content/test.csv')
val_df = pd.read_csv('/content/val.csv')
```

```python
train_df
```

```python
def split(df):
  X = df.drop(columns=['price'])
  y = df['price']
  return X,y
```

```python
X_train, y_train = split(train_df)
X_val, y_val = split(val_df)
X_test, y_test = split(test_df)
```

```python
num_cols = ['duration', 'total_stops']
datim_cols = ['date_of_journey','arrival_time','dep_time']
cat_cols = [col for col in X_train.columns if (col not in num_cols) and (col not in datim_cols)]
```

```python
num_trns = Pipeline(steps=[
                    ('imputer',SimpleImputer(strategy='mean')),
                    ('scaler',StandardScaler())
])

cat_trns = Pipeline(steps=[
                    ('imputer',SimpleImputer(strategy='most_frequent')),
                    ('encoder',OneHotEncoder(sparse_output = False,handle_unknown='ignore'))
])

date_trns = Pipeline(steps=[
                    ('imputer',SimpleImputer(strategy='most_frequent')),
                    ('date_features',DatetimeFeatures(features_to_extract=['month','year','day_of_month'],format='mixed')),
                    ('scaler',StandardScaler())
])

time_trns = Pipeline(steps=[
                    ('imputer',SimpleImputer(strategy='most_frequent')),
                    ('date_features',DatetimeFeatures(features_to_extract=['hour','minute'],format='mixed')),
                    ('scaler',StandardScaler())
])
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num',num_trns,num_cols),
    ('cat',cat_trns,cat_cols),
    ('date',date_trns,[datim_cols[0]]),
    ('time',time_trns,[datim_cols[1], datim_cols[2]])
])
```

```python
[datim_cols[1:]]
```

```python
preprocessor.fit_transform(X_train)
```

```python
algorithms = {
    'linear' : LinearRegression(),
    'svr' : SVR(),
    'decision_tree' : DecisionTreeRegressor(),
    'random_forest' : RandomForestRegressor(),
    'xgboost' : XGBRegressor()
}
```

```python
data = pd.concat([train_df,val_df],axis=0)
```

```python
X_data, y_data = split(data)
```

```python
def plot_curves(sizes,mean_scores,std_scores,label,ax):
  ax.plot(sizes,mean_scores,marker = 'o',label=label)
  ax.fill_between(sizes,mean_scores+std_scores,mean_scores-std_scores,alpha=0.5)
```

```python
def plot_learning_curves(name,algorithms,figsize=(12,4)):
  model = Pipeline(steps = [
      ('preprocessor',preprocessor),
      ('alg',algorithms)
  ])

  train_sizes,train_scores,test_scores = learning_curve(
      estimator=model,
      X=X_data,
      y=y_data,
      cv=3,
      scoring = 'r2',
      n_jobs=-1,
      random_state=42)

  mean_train_scores = np.mean(train_scores,axis=1)
  std_train_scores = np.std(train_scores,axis=1)
  train_score = f'{mean_train_scores[-1]:.2f} +/- {std_train_scores[-1]:.2f}'


  mean_test_scores = np.mean(test_scores,axis=1)
  std_test_scores = np.std(test_scores,axis=1)
  test_score = f'{mean_test_scores[-1]:.2f} +/- {std_test_scores[-1]:.2f}'


  fig,ax = plt.subplots(figsize=figsize)
  plot_curves(train_sizes,mean_train_scores,std_train_scores,f'Train({train_score})',ax)
  plot_curves(train_sizes,mean_test_scores,std_test_scores,f'Test({test_score})',ax)

  ax.set(xlabel='Training Set Size',ylabel='R2 Score',title=f'{name.title()} Learning Curves')
  ax.legend(loc='best')
  plt.show()
```

```python
for name,alg in algorithms.items():
  plot_learning_curves(name,alg)
```

```python
model = Pipeline(steps = [
    ('preprocessor',preprocessor),
    ('alg',RandomForestRegressor())
])
```

```python
model.fit(X_data,y_data)
```

```python
def evaluate_model(model,X,y):
  y_pred = model.predict(X)
  r2 = r2_score(y,y_pred)
  return r2
```

```python
print(f"r2_score on training data : {evaluate_model(model,X_data,y_data)}")
print(f"r2_score on testing data : {evaluate_model(model,X_test,y_test)}")
```
